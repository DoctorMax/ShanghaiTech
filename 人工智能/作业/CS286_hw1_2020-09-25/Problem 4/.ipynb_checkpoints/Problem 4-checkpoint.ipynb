## Problem 4. Cancer Data Analysis (30 points)
Comprehensive breast cancer risk prediction models enable identifying and targeting women at high-risk, while reducing interventions in those at low-risk. Machine learning (ML) offers an alternative approach to standard prediction modeling that may address current limitations and improve accuracy of those tools.

### Data Source 
This is an analysis of the Breast Cancer Wisconsin (Diagnostic) Data Set, obtained from UCI Machine Learning Repository: https://archive.ics.uci.edu/ml/datasets/Breast+Cancer+Wisconsin+(Diagnostic).  The data set contains 569 observations across 32 attributes. The attribute information includes ID number, Diagnosis (M = malignant, B = benign) and 30 real-valued numerical variables derived from ten features that are computed for each cell nucleus, as follows: 

![avatar](1.png#pic_left)

### Requirements (each of the following 5 tasks carries 6 points):
1. Delete redundant information (ID number);
2. Fill missing values;
3. Process non-numerical features (In the column of Diagnosis, M => 0  B => 1);
4. Do a 5-fold Cross Validation;
5. Visualize the prediction results, by drawing the ROC curve and printing the AUC value.

```python
import sklearn
import numpy as np
import pandas as pd
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestClassifier
import matplotlib.pyplot as plt
from sklearn.metrics import plot_roc_curve
from sklearn.model_selection import train_test_split
```

```python
dataset=pd.read_csv('./breast_cancer.csv')
dataset.head()
```

```python
dataset=dataset.drop(columns = ['id'])
dataset.head()
```

```python
dataset.info()
```

```python
dataset.describe()
```

```python
dataset.isnull().sum()
```

```python
dataset['concavity_mean'].fillna(dataset['concavity_mean'].median(), inplace=True)
dataset['concave points_mean'].fillna(dataset['concave points_mean'].median(), inplace=True)
dataset['concavity_se'].fillna(dataset['concavity_se'].median(), inplace=True)
dataset['concave points_se'].fillna(dataset['concave points_se'].median(), inplace=True)
dataset['concavity_worst'].fillna(dataset['concavity_worst'].median(), inplace=True)
dataset['concave points_worst'].fillna(dataset['concave points_worst'].median(), inplace=True)
```

```python
dataset.isnull().sum()
```

```python
dataset['diagnosis'][dataset['diagnosis']=='M']=0
dataset['diagnosis'][dataset['diagnosis']=='B']=1
```

```python
k_fold = KFold(n_splits=5, shuffle=True, random_state=0)
```

```python
data_train=dataset.iloc[:,1:]
data_train.apply(lambda x: (x - np.min(x)) / (np.max(x) - np.min(x)))  #pandas 标准化方法
data_train=np.array(data_train)
# data_train=np.array(dataset.iloc[:,1:])
label_train=np.array(dataset['diagnosis'],dtype=np.int32)
# data_train=dataset.iloc[:,1:]
# label_train=dataset['diagnosis']



# # print(dataset)
# print(data_train)
# print(label_train)
# print(data_train.shape)
# print(label_train.shape)
# print(type(label_train))
# print(type(label_train[0]))
# # numpy.float64
# print(type(data_train[0,0]))
```

```python
clf = RandomForestClassifier(n_estimators=9)
X_train, X_test, y_train, y_test = train_test_split(data_train, label_train, random_state=42)
clf.fit(X_train, y_train)
score = cross_val_score(clf, X_train, y_train, cv=k_fold, n_jobs=1, scoring='accuracy')
print(score)
clf_disp = plot_roc_curve(clf, X_test, y_test)
plt.show()
```
